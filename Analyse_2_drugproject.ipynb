```python
import pandas as pd

file = "su-d-19.02.01.02.01_2000.xlsx"

sheets = pd.read_excel(
    file,
    sheet_name=None,
    header=5
)

dfs = []

for year, sheet in sheets.items():

    sheet = sheet.copy()

    # nur echte Datenzeilen
    sheet = sheet[sheet["Total"].notna()]

    # Jahr hinzufügen
    sheet["Year"] = int(year)

    dfs.append(sheet)

df = pd.concat(dfs, ignore_index=True)
```

```python
counts = (
    df_all.groupby(["Year", "Offence"])
    .size()
    .reset_index(name="n")
)

duplicate_offences = counts[counts["n"] > 1]["Offence"].unique()

df_unique = df_all[
    ~df_all["Offence"].isin(duplicate_offences)
].copy()
```

```python
## Welche Straftaten haben sich wie entwicklet seit 2009? 

latest = (
    df_unique[df_unique["Year"] == 2025]
    [["Offence", "Offences_total"]]
)

oldest = (
    df_unique[df_unique["Year"] == 2009]
    [["Offence", "Offences_total"]]
)

compare = latest.merge(
    oldest,
    on="Offence",
    suffixes=("_2025", "_2009")
)

compare["Change_%"] = (
    (compare["Offences_total_2025"] - compare["Offences_total_2009"])
    / compare["Offences_total_2009"]
) * 100

compare.sort_values("Change_%").round(1)
```

```python
# Der Rückgang scheint vor allem bei Eigenkonsum-Delikten stattzufinden. 
# Nicht unbedingt im selben Ausmass Rückgang beim Handel. Aber auch Rückgang! 
```

```python
compare["Absolute_change"] = (
    compare["Offences_total_2025"]
    - compare["Offences_total_2009"]
)

compare.sort_values("Absolute_change")
```
