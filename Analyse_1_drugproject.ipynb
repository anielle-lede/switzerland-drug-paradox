```python
# Charts ausprobieren für Datawrapper
```

```python
# 0. cot weglassen
df_drugs = df[df["Metabolite"] != "cot"]

# 1. Durchschnitt pro Droge und Land berechnen
country_drug = (
    df_drugs.groupby(["Metabolite", "Country"], as_index=False)
      .agg(
          avg_daily_mean=("Daily mean", "mean"),
          n_sites=("Site ID", "nunique")
      )
)

# 2. Ranking pro Droge berechnen
country_drug["rank"] = (
    country_drug.groupby("Metabolite")["avg_daily_mean"]
    .rank(method="dense", ascending=False)
    .astype(int)
)

# 3. Führendes Land pro Droge
leaders = (
    country_drug.sort_values(["Metabolite", "avg_daily_mean"], ascending=[True, False])
    .groupby("Metabolite")
    .first()
    .reset_index()
    [["Metabolite", "Country", "avg_daily_mean"]]
    .rename(columns={
        "Country": "Leading country",
        "avg_daily_mean": "Top value"
    })
)

# 4. Schweiz-Werte
switzerland = (
    country_drug[country_drug["Country"] == "CH"]
    [["Metabolite", "rank", "avg_daily_mean", "n_sites"]]
    .rename(columns={
        "rank": "Switzerland rank",
        "avg_daily_mean": "Switzerland value",
        "n_sites": "Swiss sites"
    })
)

# 5. Zusammenführen
dw_table = switzerland.merge(leaders, on="Metabolite")

# 6. Anzahl Länder pro Droge hinzufügen
country_counts = (
    country_drug.groupby("Metabolite")["Country"]
    .nunique()
    .reset_index()
    .rename(columns={"Country": "Number of countries"})
)

dw_table = dw_table.merge(country_counts, on="Metabolite")

# 7. Spalten sortieren und runden
dw_table = dw_table[
    [
        "Metabolite",
        "Switzerland rank",
        "Number of countries",
        "Switzerland value",
        "Leading country",
        "Top value",
    ]
]

dw_table["Switzerland value"] = dw_table["Switzerland value"].round(2)
dw_table["Top value"] = dw_table["Top value"].round(2)

# 8. Anzeigen
dw_table.sort_values("Switzerland rank")
```

```python
dw_table = dw_table.rename(columns={
    "Metabolite": "Drug",
    "Switzerland rank": "Swiss rank",
    "Number of countries": "Countries analysed",
    "Switzerland value": "Swiss average",
    "Leading country": "Top country",
    "Top value": "Highest average"
})
```

```python
dw_table.sort_values("Swiss rank")
```

```python
dw_table.to_csv("switzerland_drug_ranking_datawrapper.csv", index=False)
```

```python
# Top 5 Städte pro Droge mit den weekendspikes / Which five cities show the strongest weekend spike for each drug?

# Stadtnamen bereinigen
df["City clean"] = (
    df["City"]
    .str.replace(r"\s*\(\d+.*?\)", "", regex=True)
    .str.strip()
)

# Durchschnittlicher Wochenend-Effekt pro Stadt
weekend_city = (
    df.groupby(
        ["Metabolite", "City clean", "Country"],
        as_index=False
    )["Weekend increase %"]
    .mean()
)

# Top 5 Städte pro Droge
top5_weekend = (
    weekend_city
    .sort_values(
        ["Metabolite", "Weekend increase %"],
        ascending=[True, False]
    )
    .groupby("Metabolite")
    .head(5)
)

top5_weekend
```

```python
#In welcher Schweizer Stadt steigt Kokain, MDMA oder Cannabis am Wochenende am stärksten an? 

ch_weekend = df[df["Country"] == "CH"]

top_weekend_ch = (
    ch_weekend.loc[
        ch_weekend.groupby("Metabolite")["Weekend increase %"].idxmax()
    ][["Metabolite", "City", "Weekend increase %"]]
    .sort_values("Weekend increase %", ascending=False)
)

top_weekend_ch
```
